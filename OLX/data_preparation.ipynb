# Trabalhando na coluna valor
Objetivo é deixar essa coluna em formato int e dropar valores estranhos

```python
# tirando o simbolo do reais
val = dados['VALOR'].str.split(' ', expand=True)
dados['VALOR'] = val[1]
dados
```

```python
#suprimindo todos os pontos
a = dados['VALOR']
lista = [ str(i).replace('.' ,'') for i in  a]
dados['VALOR'] = lista
dados.head()
```

```python
#suprimindo valores em formato de parcela ou seja que contem um "x"
dados2 = dados[dados['VALOR'].str.contains('x')== False]
dados2.head(20)
```

```python
# agora vamos colocar a coluna em formato int
dados2['VALOR'] = dados2['VALOR'].astype('int')
dados2.info()
```

```python
dados2 = dados2[(dados2['VALOR'] > 500) & (dados2['VALOR'] <= 15000) ]
dados2
```

```python
dados2['LOCAL'] = dados2['LOCAL'].str.lower()
a = dados2['LOCAL'].str[-2:]
dados2['Estado'] = a
dados2

```

```python
dados['TITULO'] = dados['TITULO'].str.lower()
dados.info()
```

```python
#marcas = ["samsung", "xiaomi", "apple", "iphone" , "motorola", "asus", "lg", "huawei", "lenovo", "nokia"]
lista = []
for i in dados["TITULO"]:
    if "iphone" in i or "apple" in i:
       lista.append("Iphone")
    elif "samsung" in i:
        lista.append("Samsung")
    elif "xiaomi" in i:
        lista.append("Xiaomi")
    elif "lg" in i:
        lista.append("Lg")
    elif "motorola" in i or "moto" in i :
        lista.append("Motorola")
    elif "nokia" in i:
        lista.append("Nokia")
    elif "asus" in i:
        lista.append("Asus")
    elif "huawei" in i:
        lista.append("Huawei")
    elif "lenovo" in i:
        lista.append("Lenovo")
    else:
        lista.append("Other")
dados["marcas"] = pd.DataFrame(lista)
dados["TITULO"].shape
```
